# 45 - Triangular, Pentagonal, and Hexagonal

## Problem Statement

Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

\begin{align*}
\textbf{Triangle} \quad & T_n = \frac{n(n+1)}{2} & \text{Examples: } 1, 3, 6, 10, 15, \ldots \\
\textbf{Pentagonal} \quad & P_n = \frac{n(3n-1)}{2} & \text{Examples: } 1, 5, 12, 22, 35, \ldots \\
\textbf{Hexagonal} \quad & H_n = n(2n-1) & \text{Examples: } 1, 6, 15, 28, 45, \ldots
\end{align*}

It can be verified that $T_{285} = P_{165} = H_{143} = 40755$.

Find the next triangle number that is also pentagonal and hexagonal.

## Solution

$T_n = \frac{n(n + 1)}{2}, \quad n = \left\lfloor \frac{-1 + \sqrt{1 + 8x}}{2} \right\rfloor$

$P_n = \frac{3n^2 - n}{2}, \quad n = \left\lfloor \frac{1 + \sqrt{1 + 24x}}{6} \right\rfloor$


```python
import math

def is_triangular(x):
    y = (-1 + math.sqrt(1 + 8 * x)) / 2
    return y == int(y)

def is_pentagonal(x):
    y = (1 + math.sqrt(1 + 24 * x)) / 6
    return y == int(y)

n = 144
while True:
    x = n * (2 * n - 1)
    if is_pentagonal(x) and is_triangular(x):
        res = x
        break
    n += 1

res
```
